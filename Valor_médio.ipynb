```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2 = pd.read_csv(path + "/player_market_values.csv")
df2
```

```python
gdp_df = pd.read_csv('/content/gdp-per-capita-worldbank.csv')
display(gdp_df.head())
```

# Análise entre o valor médio dos jogadores e o PIB per capita medio global

```python
player_market_value_per_year = df2.groupby('year')['market_value_eur_m'].mean().reset_index()
player_market_value_per_year.rename(columns={'market_value_eur_m': 'valor médio de mercado dos jogadores(em milhoes de euros)', 'year':'ano'}, inplace=True)
display(player_market_value_per_year.head())
```

```python
gdp_per_capita_per_year = gdp_df.groupby('Year')['GDP per capita'].mean().reset_index()
gdp_per_capita_per_year.rename(columns={'Year': 'ano', 'GDP per capita': 'PIB per capita médio'}, inplace=True)
display(gdp_per_capita_per_year.head())
```

```python
merged_df = pd.merge(player_market_value_per_year, gdp_per_capita_per_year, on='ano', how='inner')
display(merged_df.head())
```

```python
fig, ax1 = plt.subplots(figsize=(12, 7))

color = 'tab:red'
ax1.set_xlabel('Ano')
ax1.set_ylabel('Valor de Mercado Médio (milhões de euros)', color=color)
sns.lineplot(x='ano', y='valor médio de mercado dos jogadores(em milhoes de euros)', data=merged_df, ax=ax1, color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instanciar um segundo eixo que compartilha o mesmo eixo x
color = 'tab:blue'
ax2.set_ylabel('PIB per Capita Médio', color=color)
sns.lineplot(x='ano', y='PIB per capita médio', data=merged_df, ax=ax2, color=color, marker='x')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Comparação entre Valor de Mercado Médio de Jogadores e PIB per Capita Médio Global ao Longo do Tempo')
fig.tight_layout()  # para garantir que não haja sobreposição
plt.show()
```

Linha vermelha: Valor médio de mercado dos jogadores
Em 2010, o valor médio era de aproximadamente 8 milhões de euros.
Entre 2010 e 2019, observa-se um crescimento praticamente contínuo.
Entre 2020 e 2021, o crescimento desacelera, mas os valores continuam aumentando.
Em 2022 e 2023, ocorre uma nova valorização significativa, atingindo cerca de 59 milhões de euros.
Em 2024, há uma pequena redução em relação a 2023, mas o valor permanece muito elevado.

No período completo, o valor médio dos jogadores aumentou mais de sete vezes.

Linha azul: PIB per capita médio
O PIB per capita cresce de forma relativamente constante entre 2010 e 2019.
Em 2020, observa-se uma queda acentuada, que provavelmente está relacionada aos impactos econômicos da pandemia de COVID-19.
A partir de 2021, inicia-se uma recuperação.
Entre 2022 e 2024, o indicador volta a crescer, alcançando seu maior valor no final da série.

O gráfico evidencia que ambas as variáveis apresentam tendência de crescimento no longo prazo, mas o valor médio de mercado dos jogadores cresce muito mais rapidamente que o PIB per capita médio global. Além disso, enquanto o PIB per capita sofreu uma queda significativa em 2020 devido à crise econômica mundial, o mercado de jogadores mostrou maior resiliência e continuou sua trajetória de valorização. Portanto, embora exista uma evolução semelhante no longo prazo, não é possível concluir apenas com esse gráfico que há uma relação de causa e efeito entre as duas variáveis; seriam necessárias análises estatísticas, como o cálculo da correlação ou uma regressão, para investigar essa relação de forma mais rigorosa.
