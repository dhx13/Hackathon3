## Limpeza Inicial football

```python
club_fin = pd.read_csv(r'football\club_financials.csv')
leagues = pd.read_csv(r'football\league_metrics.csv')
player_values = pd.read_csv(r'football\player_market_values.csv')
player_full = pd.read_csv(r'football\players.csv', engine='python', on_bad_lines='skip')
records = pd.read_csv(r'football\record_transfers.csv')
all_transfers = pd.read_csv(r'football\transfers_history.csv')

# Check if players in player_values are also in player_full, if so add a country collum to player_values of the players country of origin
player_values = player_values.merge(
    player_full[['name', 'country_of_birth']],
    how='left',
    left_on='player_name',
    right_on='name'
)
player_values.drop(columns='name', inplace=True)

# Arrumar o país de origem dos jogadores que não tem, a partir do dicionário de jogadores e países
jogadores_paises = {
    "Luis Suarez": "Uruguay",
    "Luka Modric": "Croatia",
    "Mesut Ozil": "Germany",
    "Vinicius Jr": "Brazil",
    "Sadio Mane": "Senegal",
    "Aurelien Tchouameni": "France",
    "Moises Caicedo": "Ecuador",
    "Pau Cubarsi": "Spain",
    "Antonio Rudiger": "Germany",
    "Joao Cancelo": "Portugal",
    "Theo Hernandez": "France",
    "Andy Robertson": "Scotland",
    "Alisson Becker": "Brazil",
    "Andre Onana": "Cameroon",
    "Warren Zaire-Emery": "France",
    "Arda Guler": "Turkey",
    "Fermin Lopez": "Spain",
    "Joao Neves": "Portugal",
    "Estevão Willian": "Brazil",
    "Martin Odegaard": "Norway",
    "Lautaro Martinez": "Argentina",
    "Dusan Vlahovic": "Serbia",
    "Rasmus Hojlund": "Denmark",
    "Niklas Sule": "Germany"
}

for player, country in jogadores_paises.items():
    player_values.loc[player_values['player_name'] == player, 'country_of_birth'] = country

# # Mudar England para United Kingdom, pois é o nome que está na lista de países do macrodata
player_values.loc[player_values['country_of_birth'] == 'England', 'country_of_birth']  = 'United Kingdom'
player_values.loc[player_values['country_of_birth'] == 'Scotland', 'country_of_birth']  = 'United Kingdom'
player_values.loc[player_values['country_of_birth'] == 'Korea, South', 'country_of_birth']  = 'South Korea'
player_values.loc[player_values['country_of_birth'] == 'East Germany (GDR)', 'country_of_birth']  = 'Germany'

# Ver se todos os países dos jogadores estão na lista de países do macrodata e imprimir alguns jogadores que não estão na lista de países do macrodata
# paises_macrodata = countries['Country Name'].unique()
# total_jogadores = player_values.shape[0]
# jogadores_com_pais = player_values['country_of_birth'].isin(paises_macrodata).sum()
# jogadores_sem_pais = player_values[~player_values['country_of_birth'].isin(paises_macrodata)]
# print(jogadores_sem_pais)

player_values['year_of_birth'] = player_values['year']-player_values['age']

player_values.to_csv(r'football\new_player_values.csv', index=False)
```
